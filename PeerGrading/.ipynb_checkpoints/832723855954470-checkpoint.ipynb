## 1. Wattz-Strogatz networks


* **Use** `nx.watts_strogatz_graph` **to generate 3 graphs with 500 nodes each, average degree = 4, and rewiring probablity  p=0,0.1, and 1.**

We define three neetworks and store them in a list. 

```python
A = nx.watts_strogatz_graph(500, 4, p=0)
B = nx.watts_strogatz_graph(500, 4, p=0.1)
C = nx.watts_strogatz_graph(500, 4, p=1)
graphs = [A, B, C]
```

* **Calculate the average shortest path length  ⟨d⟩  for each one.**

This is done using the `networkx.average_shortest_path_length()`

```python
print("Average shortest path length:")
for g, p in zip(graphs, [0, 0.1, 1]):
    print("\t For p={0} : {1}".format(p, nx.average_shortest_path_length(g)))
```

The model starts as a ring of nodes (see [chapter 3](http://networksciencebook.com/chapter/3#clustering-3-9)). Clearly, when p=0 each node is only connected to its close neighbours in the ring of nodes. This results in high clustering and therefore also high average shortest path length. This makes a lot of sense, since the shortest path to nodes on the other sider of the ring, will be quite high. 

As p increases, the average shortest path will decrease since each node might be connected to nodes that are on the opposite site of the ring.

* **Describe what happens to the network when p=1.**

With p=1, each link in the network will be rewired to a randomly chosen node. This creates a random network structure, with low average short path length and low clustering. 

* **Generate 50 networks with  N=500 ,  ⟨k⟩=4 , for each of  p={0,0.01,0.03,0.05,0.1,0.2} .**

We essentielly do two for-loops and store the networks in a list of lists.

```python
nr_networks = 50
N = 500
k = 4
p = [0, 0.01, 0.03, 0.05, 0.1, 0.2]

networks = [[] for _ in p]

for i, p_ in enumerate(p):
    networks[i] = [nx.watts_strogatz_graph(N, k, p_) for _ in range(nr_networks)]
```

* **Calculate the average of  ⟨d⟩  as well as the standard deviation over the 50 networks**

We again use the `networkx.average_shortest_path_length()` class and store the average shortest path lengths in a numpy array with same dimensions as the networks list of lists data structure. 

```python
network_d_avgs = np.zeros((len(p), nr_networks))

for i, networks_list in enumerate(networks):
    for j, g in enumerate(networks_list):
        network_d_avgs[i, j] = nx.average_shortest_path_length(g)
```

Now, we can easily calculate the mean and standard deviasion using the numpy buildin functions.

```python
means = np.mean(network_d_avgs, axis=1)
stds = np.std(network_d_avgs, axis=1)
```

* **create a plot that shows how the path length decreases very quickly with only a little fraction of re-wiring. Use the standard deviation to add errorbars to the plot.**

```python
plt.figure(figsize=(10,5))
plt.errorbar(p, means, stds, fmt='-o')
plt.xlabel("p")
plt.ylabel("ave(<d>)")
plt.title("Average shortest path length with different p")
plt.show()
```

Clearly, the path length decreases very quickly as p increases.

1. For p=0, there is no standard deviation. The reason lies in the definition of the Watts-Strogatz model. The network is initialized as a ring of nodes, where each node is connected to its neighbours, depending on the average degree chosen to initialize the network. We rewire each link with probability p, so if p=0, we do not rewire any links. Hence the initialized network, is also the resulting network, which explains the no standard deviation.

2. As p increases (excluding p=0), the standard deviation decreases. This is due to the fact that for each node, there is a higher probability for the node connecting to a node on the opposite side of the "ring".  

## 2. The Barabasi-Albert Model

* **Create a 100 node BA network using a BA model that you've coded on your own (so don't use the built-in NetworkX function, but the one you created during week 3).**

The network is coded as its own class with all functions needed for Part 2 and Part 3 exercises. It stores the graph in a python dictionary, which makes it easy to convert the graph to a networkx graph. The probability of connecting a new node to node i is calculated by $p_i = \frac{k_i}{\sum_j k_j}$.

```python
class BarabasiAlbertNetwork:
    graph = {}

    def __init__(self, N, seed=1337):
        self.graph = {}
        # Add seed for reproducability
        np.random.seed(seed)
        self.N = N
        
        self.generate_model()
       
    def generate_model(self):
         # Add first two nodes and create a link between them
        self.add_node(0)
        self.add_node(1)
        self.add_edge(0, 1)

        # generate all N nodes
        for t in range(2, self.N):

            # Init probability vector for all nodes in graph
            p_vector = np.zeros(len(self.graph))

            k_sum = 0
            for i in range(0, len(p_vector)):
                # k_i is node degree 
                k_i = float(len(self.graph[i]))
                k_sum += k_i
                p_vector[i] = k_i
            
            # Normalize probability vector, so it sums to 1
            p_vector /= k_sum
            
            # Choose which node to connect with which with probability proportional to node degree
            edge = np.random.choice(np.arange(len(p_vector)), size=1, p=p_vector)
            
            # Add the new node with the edge from random.choice
            self.add_node(t)
            self.add_edge(t, edge[0])
                
        print("Model generated!")
        
    def add_node(self, value):
        if value not in self.graph:
            self.graph[value] = []
        else:
            print("Tried adding node twice with value: {0}".format(value))

    def add_edge(self, node_value_1, node_value_2):
        self.graph[node_value_1].append(node_value_2)
        self.graph[node_value_2].append(node_value_1)

    def export_to_nx(self):
        G = nx.Graph()
        G.add_nodes_from(self.graph.keys())

        for k, v in self.graph.items():
            G.add_edges_from(([(k, t) for t in v]))

        return G
    
    def get_max_degree(self):
        return max([len(connections) for connections in self.graph.values()])
    
    def get_min_degree(self):
        return min([len(connections) for connections in self.graph.values()])
    
    def get_all_degrees(self):
        return [len(connections) for connections in self.graph.values()]
    
    def get_degree_of(self, node):
        return len(self.graph[node])
    
    def get_neighbours_degrees(self, node):
        return [len(self.graph[neighbour]) for neighbour in self.graph[node]]
```

Now, we can simply call initialize an object of the class with N=100.

```python
barabosi_albert_model = BarabasiAlbertNetwork(N=100)
```

* **Plot the network using NetworkX.**

The following function is created in case we want to plot more graphs. It scales the nodes according to how many connections they have.

```python
def plot_barabosi_albert_graph(graph, title):
    d = dict(nx.degree(graph))
    nx.draw(graph, nodelist=d.keys(), node_size=[v * 60 for v in d.values()])
    plt.title(title)
    plt.show()
```

```python
plt.figure(figsize=(10,5))
G = barabosi_albert_model.export_to_nx()
plot_barabosi_albert_graph(G, "500 node Barabosi-Albert netork")
```

Clearly, the network has few hubs. A single hub is clearly seen as the most connected, which is most likely one of the first two nodes added to the network. This is due to how the model adds an edge. It simply makes a random choice with probabilities proportional to the current node degrees.

* **Bin the degree distribution**

I choose to bin the degree distribution using `collections.Counter` class.

Note, according to [chapter2](http://networksciencebook.com/chapter/2#degree) (and every book on probability theory! :b ) the degree distribution is normalized.

```python
degree_dict = dict(map(lambda kv: (kv[0], kv[1] / 5000.0), degree_dict.iteritems()))
```

* **Plot the distribution. Plot it with both linear and log-log axes.**

The plotting is defined as a function, since it will be used in later exercises. 

```python
def plot_degree_distributions(degree_dictionary, title):
    f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,7))

    # First plot
    f.suptitle(title, fontsize=16)    
    ax1.set_ylabel("$p_k$")
    ax1.scatter(degree_dictionary.keys(), degree_dictionary.values(), marker=".")

    # Second plot
    ax2.set_yscale('log')
    ax2.set_xscale('log')
    ax2.set_xlabel("k")
    ax2.set_ylabel("$p_k$")
    ax2.scatter(degree_dictionary.keys(), degree_dictionary.values(), marker=".")
    plt.show()
```

```python
plot_degree_distributions(degree_dict, "Degree distribution with different axes")
```

The degree distribution clearly shows that the implemented barabosi-albert network is a scale-free network as we see that $\log p_k$ more or less depends linearly on $\log k$, which is expected in scale-free networks, but not in random networks. 

## 3. Power-laws and the friendship paradox

**Explore the friendship paradox on the 5000 node Barabossi network.** 

* **Pick a node _i_ at random (e.g. use `random.choice`). [Find its degree]**(http://networkx.lanl.gov/reference/generated/networkx.Graph.degree.html).
* **Find _i_'s [neighbors](http://networkx.lanl.gov/reference/generated/networkx.Graph.neighbors.html). And calculate their average degree.**
* **Compare the two numbers to check if it's true that _i_'s friends (on average) have more friends than _i_.**
* **Do this 1000 times. How many out of those 1000 times is the friendship paradox true?**

All of the above can be defined in a single for-loop, with the already implemented functions of the `BarabasiAlbertNetwork` class.

```python
friendship_paradox_true = np.zeros(1000)

for i in range(1000):
    node = np.random.choice(np.arange(5000))
    node_degree = barabosi_albert_model_5000.get_degree_of(node)
    neighbour_degrees = barabosi_albert_model_5000.get_neighbours_degrees(node)
    avg_neighbour_degree = sum(neighbour_degrees) / len(neighbour_degrees)
    friendship_paradox_true[i] = node_degree < avg_neighbour_degree

print("Friendship paradox true in {0} out of 1000 trials".format(np.sum(friendship_paradox_true)))
```

Hence the friendship paradox is true in approximately 84% of the 1000 trials. If a hub is chosen in the random choice of node, then the average of its neighbours will most likely not be higher. However, if a node, which is not a hub is chosen, then the average will most likely be higher, since there is a high probability of the node being connected to a hub.

* **Create simple network statistics for the 113th house of representatives**.

First, the data is read into a dataframe. Furthermore, a function used to generate filename from the wikinames in the df is defined. 

```python
df = pd.read_csv("../files/data_US_congress/H113.csv")
path = "./data/"
generate_filename = lambda x: path + x + ".txt"
```

Next, we define functions needed to create the network from the extracted text files. The regex can return two types of links, links where the name of the wikipage is specified and links of type $\text{<wikipagename>} | \text{<defined link placeholder on wikipage>}$. For the ladder case, a simple `str.split()` function is used to get the actual wikipage link. 

```python
# Regex expression
match_link = "\[\[(.*?)\]\]"

# Split URL and take first part, which corresponds to the exact wikipage link
extract_link = lambda x: x.split("|")[0]
```

The wikinames which we have available contains _ instead of space, hence this is substituted.

```python
# Use _ instead of " " to allow matching
replace_spaces = lambda x: x.replace(" ", '_')
```

To make things even more complicated, there is a special case with these links since a wikipage can have multiple redirect links. The links we extract might be one of those redirect links, that do not match the exact wikipagename we have available, but it is still a valid connection. We should not lose this information, hence we get all possible redirects for each of the wikipage of the members of the 113th congress and match based on these instead of just the actual wikipagename. This can be done by utilizing the wikipedia API module [backlinks](https://www.mediawiki.org/wiki/API%3aBacklinks). 

We first define a function to generate backlinks query based a wikipage title. 

```python
# Special case of other links with different wiki link that redirects to one of our wikipages
def query_gen(title):
    base = "http://en.wikipedia.org/w/api.php?"
    action = "action=query"
    list_ = "list=backlinks"
    filter_ = "blfilterredir=redirects"
    bltitle= "bltitle=" + title
    bllimit = "bllimit=max"
    dataformat = "format=json"
    return "%s%s&%s&%s&%s&%s&%s" % (base, action, list_, filter_, bltitle, bllimit, dataformat)
```

The response is in json format and can be loaded into a python dictionary using `json.loads()`. The following function extracts the title of the redirect links based on the json response.

```python
def get_alternative_links(response):
    json_dict = json.loads(response)
    return map(lambda x: x["title"], json_dict["query"]["backlinks"])
```

Now, we loop through the dataframe. For each row, we call the wikipedia API, extract the redirect links and store all wikipage links in a dictionary, with the corresponding actual wikipagename that we have available. We furthermore add the actual wikipagenames since they are not included in the API responses. 

```python
redirects_to_actual_wikiname = {}
for index, row in df.iterrows():
    query = query_gen(row["WikiPageName"])
    response = urllib2.urlopen(query).read()
    all_redirects = get_alternative_links(response)
    
    # Replace spaces with _ 
    all_redirects = list(map(replace_spaces, all_redirects))
    
    for redirect in all_redirects:
        redirects_to_actual_wikiname[redirect] = row["WikiPageName"]
    
    redirects_to_actual_wikiname[row["WikiPageName"].decode('utf-8')] = row["WikiPageName"]
```

To check whether a link is a link to a politicians wikipage, we can simply check (in constant time!) if it is in the python dictionary. 

```python
is_politician_in_113_congress_filter = lambda link: link in redirects_to_actual_wikiname
```

We store the graph information (nodes, links and attributes) in two python dictionaries, since they work well with the `networkx` library. We do this by looping over the pandas dataframe and applying all of the previous specified functions in correct order. Since the suggested `DiGraph` does not allow multilinks, we use `set` on the matched links. 

```python
links_dict = {}
attributes_dict = {}

for index, row in df.iterrows():
    name = row["WikiPageName"]
    text = io.open(generate_filename(name), encoding="utf-8").read()
    
    # Find all links in text
    all_links = re.findall(match_link, text)
    
    # apply the specified functions above in correct order
    all_links = map(lambda link: redirects_to_actual_wikiname[link],#lambda function which gets the correct wikipagename
                    filter(is_politician_in_113_congress_filter,
                        map(replace_spaces, 
                          map(extract_link, 
                              all_links))))
    
    # Store in dict
    links_dict[name] = set(all_links)
    attributes_dict[name] = {"party": row["Party"], "state": row["State"]}
```

Now, we can simply create the `DiGraph()` by adding the nodes and edges from the dictionaries. Furthermore, see that the properties are added to the nodes as attributes, so for a given node, we have the attributes "party" and "state". 

```python
G = nx.DiGraph()

G.add_nodes_from(links_dict.keys())

# Store properties of each node as node attributes of networkx Graph
nx.set_node_attributes(G, attributes_dict)

for key, links in links_dict.items():
    G.add_edges_from([(key, link) for link in links])
```

We can plot the graph. To illustrate how attributes can be used, the colour of the nodes are determined from the party of the representative.

```python
pos = nx.spring_layout(G, k=0.15, iterations=30)

node_color = []

for node in G.nodes(data=True):
    if node[1]["party"] == "Republican":
        node_color.append("red")
    else:
        node_color.append("blue")

d = dict(nx.degree(G))
plt.figure(figsize=(10,5))
plt.title("133th congress politician network")
nx.draw(G, pos, node_size=[v * 10 for v in d.values()], node_color=node_color)
plt.show()
```

The network is quite complex and it might be hard to conclude anything very useful about the network. However, clearly we see two hubs which is not very likely for random-networks, but more likely for the scale-free networks. One hub is red and one is blue, so we have a very connected politician from both parties!
