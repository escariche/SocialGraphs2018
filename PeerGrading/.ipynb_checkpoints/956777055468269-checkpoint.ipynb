> ## 1. Watts-Strogatz Networks

> * Use nx.watts_strogatz_graph to generate 3 graphs with 500 nodes each, average degree = 4, and rewiring probablity $p = 0, 0.1, \textrm{and} 1$. Calculate the average shortest path length $\langle d \rangle$ for each one. Describe what happens to the network when $p = 1$.

```python
import networkx as nx

n = 500
k = 4
p = [0,0.1,1]

WSG0 = nx.watts_strogatz_graph(n,k,p[0])
WSG01 = nx.watts_strogatz_graph(n,k,p[1])
WSG1 = nx.watts_strogatz_graph(n,k,p[2])

aspl0 = nx.average_shortest_path_length(WSG0)
aspl01 = nx.average_shortest_path_length(WSG01)
aspl1 = nx.average_shortest_path_length(WSG1)

print("Average shortest path length for p = 0: " + str(aspl0))
print("Average shortest path length for p = 0.1: " + str(aspl01))
print("Average shortest path length for p = 1: " + str(aspl1))
```

> * Let's understand the behavior of the WS model as we increase p in more detail.. Generate 50 networks with $N = 500$, $\langle k \rangle = 4$, for each of $p = \{0, 0.01, 0.03, 0.05, 0.1, 0.2\}$. Calculate the average of $\langle d \rangle$ as well as the standard deviation over the 50 networks, to create a plot that shows how the path length decreases very quickly with only a little fraction of re-wiring. Use the standard deviation to add errorbars to the plot. My version of the plot is below (since a picture's worth 1000 words).

```python
import random 
import statistics
import numpy as np

liste = [0,0.01,0.03,0.05,0.1,0.2]
liste_d = []
avg = [0,0,0,0,0,0]
sd = [0,0,0,0,0,0]

for x in range(len(liste)):
    for i in range(50):
        temp = nx.watts_strogatz_graph(500,4,liste[x])
        liste_d.append(nx.average_shortest_path_length(temp))
    avg[x] = np.mean(liste_d)
    sd[x] = statistics.stdev(liste_d)
    liste_d = []
```

```python
plt.errorbar(liste, avg, sd, linestyle = 'None', marker='.')
plt.plot(liste, avg)
```

> ## 2. The Barabasi-Albert Model
We're going to create our own Barabasi-Albert model (a special case) in right in a notebook. Follow the recipe below for success
> 
> * Create a 100 node BA network using a BA model that you've coded on your own (so don't use the built-in NetworkX function, but the one you created during week 3). And plot it using NetworkX.

```python
import networkx
import random
import numpy

G = networkx.Graph()
G.clear()
G.add_edge(1,2)

def chooseRandomNode(edges):
    flat_list = []
    for e in edges:
        for item in e:
            flat_list.append(item)
    return random.choice(flat_list)

for i in range(100):
    G.add_edge( i, chooseRandomNode(G.edges()) )
    
        
networkx.draw(G, with_labels=False, node_size = 5, width = 1)
```

> * Now create a 5000 node network.

```python
for i in range(100,5000):
    G.add_edge( i, chooseRandomNode(G.edges()) )
    
        
networkx.draw(G, with_labels=False, node_size = 5, width = 1)
```

> * What's the maximum and minimum degree?

```python
degrees = G.degree()

deg_min = 99999999
deg_max = 0

histogram_list = []

for x in degrees:
    histogram_list.append(x[1])
    if (x[1] > deg_max):
        deg_max = x[1]
    elif (x[1] < deg_min):
        deg_min = x[1]
        
print("Minimum degree is: " + str(deg_min))
print("Maximum degree is: " + str(deg_max))
```

> * Now, bin the degree distribution, for example using numpy.histogram.


```python
v = sorted(np.unique(histogram_list))
print(np.histogram(histogram_list, bins=v))
plt.hist(histogram_list, bins=v)
plt.show()
```

> ## 3. Power-laws and the friendship paradox

>Next step is to explore the Friendship paradox. This paradox states that almost everyone have fewer friends than their friends have, on average. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution. The explanation is that almost everyone is friends with a hub, that drives up the average degree of the friends. Let's explore that in the 5000 node BA network we've just generated. Do the following:

> * Pick a node i at random (e.g. use random.choice). Find its degree.

```python
count = 0

randNum = random.choice(range(G.size()))

neighbors = list(networkx.neighbors(G,randNum))

degree = len(neighbors)

print("The degree of node", randNum,"is", degree)
```

> * Find i's neighbors. And calculate their average degree.

```python
avg_deg = networkx.average_neighbor_degree(G, nodes = neighbors)


num = 0

for k in avg_deg:
    num += avg_deg.get(k)

num = num / len(list(avg_deg))

print("The average degree of the friends of",randNum,"is",num)
```

> * Compare the two numbers to check if it's true that i's friends (on average) have more friends than i.

```python
if (num > degree):
    print("The friends of",randNum,"have, on average, more friends than",randNum)
    
else:
    print("The friends of",randNum,"do not, on average, have more friends than",randNum)
```

> * Do this 1000 times. How many out of those 1000 times is the friendship paradox true?

```python
count = 0
for i in range(1000):
    randNum = random.choice(range(G.size()))

    neighbors = list(networkx.neighbors(G,randNum))

    degree = len(neighbors)

    avg_deg = networkx.average_neighbor_degree(G, nodes = neighbors)

    num = 0

    for k in avg_deg:
        num += avg_deg.get(k)

    num = num / len(list(avg_deg))

    if (num > degree):
        count += 1

print("Out of 1000 times, the friendship paradox is true",count,"times")
```

> ## 5. Basic stats for the network

> * Create simple network statistics for the 113th house of representatives.

```python
df1['Name'] = df1.WikiPageName.apply(lambda x: x.replace('_', ' '))
import networkx as nx
G=nx.DiGraph()

names = set(df1.Name)
#names = [n.replace('_', ' ') for n in names]
for i in range(len(df1)):
    wiki_name = df1['Name'].iloc[i]
    f = io.open('wiki_data/113/'+df1['WikiPageName'].iloc[i]+'.txt', mode="r", encoding="utf-8")
    txt = f.read()
    strings = re.findall(r'\[\[(.*?)\]\]', txt)
    #check if name present in the string
    #df = pd.DataFrame({'name':strings})
    multi_list = [i.split('|')[0] for i in list(strings)]
    #flat_list = [item for sublist in multi_list for item in sublist]
    intersection_set = set(multi_list).intersection(set(names))
    intersection_list = list(intersection_set)
    for i in range(len(intersection_list)):
        G.add_edge(unicode(wiki_name,'utf-8'),intersection_list[i])
nx.draw(G, with_labels=True, font_weight='bold', node_size = 30, node_color = "#33B7FF",font_size = 3)   
```

> * What is the number of nodes in the network? And the number of links?

```python
#number of nodes
print("The number of nodes is " + str(len(G.nodes())))

#number of edges
print("The number of edges is " + str(G.number_of_edges()))
```

> * Plot the in and out-degree distributions.

```python
df_in = pd.DataFrame(dict(G.in_degree(G.nodes)).items())
df_in.columns = ['dzqkjzq', 'in-degree']
df_in = df_in.groupby(['in-degree']).size().plot('bar')
```

```python
df_out = pd.DataFrame(dict(G.out_degree(G.nodes)).items())
df_out.columns = ['dzqkjzq', 'out-degree']
df_out = df_out.groupby(['out-degree']).size().plot('bar')

```
