# Extracting data from Wikipedia's API

What is the the difference between the html page and the wiki-source?

    The main difference between them is their purpose. Since the html is going to be displayed inmediately after downloading it, it is formatted to perform best in that situation. Meanwhile, the wiki-source is ready to ease the access to the contained infomation, since it will be extracted before being displayed in another app. 

What are the various parameters you can set for a query of the wikipedia api?

    There are 14 available parameters

```python
import urllib.request

#URL components
basic = 'https://en.wikipedia.org/w/api.php'
action_search = 'action=query&list=search'
action_help = 'action=help'
topic = 'srsearch=%s' % 'Mahatma%20Gandhi'
form = 'utf8=&format=json'

search = '%s?%s&%s&%s' % (basic, action_search, topic, form)
print(search)
with urllib.request.urlopen(search) as response:
    html = response.read()
```

```python
import json
source = json.loads(html)
query = source['query']
info = query['search'][0]
print(info.keys())
info["title"]
```

```python
lookup = '%s?%s&%s&%s' % (basic, action_help, 'modules=query', form)
print(lookup)
with urllib.request.urlopen(lookup) as response:
    html2 = response.read()
print(html2)
```
