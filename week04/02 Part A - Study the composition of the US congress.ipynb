#### 113th congress

```python
cong113 = pd.read_csv('~/Google Drive/DTU/Courses/SocialGraphs/socialgraphs2018/files/data_US_congress/H113.csv')
cong114 = pd.read_csv('~/Google Drive/DTU/Courses/SocialGraphs/socialgraphs2018/files/data_US_congress/H114.csv')
cong115 = pd.read_csv('~/Google Drive/DTU/Courses/SocialGraphs/socialgraphs2018/files/data_US_congress/H115.csv')

cong113['congress_number'] = 113
cong114['congress_number'] = 114
cong115['congress_number'] = 115
```

```python
all_members = pd.concat([cong113, cong114, cong115]).reset_index(drop=True)
all_members
```

```python
groups = all_members.groupby('WikiPageName')
for k, g in groups:
    print(g)
    break
```

```python
times_in_congress = groups.size().sort_values()
result = times_in_congress.groupby(times_in_congress).size()
```

```python
import matplotlib.pylab as plt
result.plot('bar')
plt.show()
```
